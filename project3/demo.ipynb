```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
input_text = "dog"
```

```python
embeddings = hf_embeddings.embed_documents(input_text)
```

```python
print(embeddings)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python

text1 = ["pen","cat"]
text2 = "dog"
```

```python
embedding1 = model.encode(text1, convert_to_tensor=True)
embedding2 = model.encode(text2, convert_to_tensor=True)
```

```python
#using sentence-transforms

similarity = util.pytorch_cos_sim(embedding1, embedding2)
# print(f"Similarity between {text1},{text2}:", similarity.item())
# print(similarity)

for idx, text in enumerate(text1):
    print(f"Similarity between '{text}' and '{text2}': {similarity[idx][0].item()}")
```

```python
#using huggingFaceEmbedding

similar = util.pytorch_cos_sim(embeddings, embedding1[0])

print(f"Similarity between {input_text},{text1[0]}:", similar.item())
```
